### Remove outliters
Outliers are data points that are significantly different from the rest of the dataset. They can distort analysis, affect mean and standard deviation, and lead to misleading results. Outliers are usually detected using methods like

```python
# Find the first quartile (25%) of the Payload Mass column
Q1 = df['Payload Mass (kg)'].quantile(0.25)
# Find the third quartile (75%) of the Payload Mass column
Q3 = df['Payload Mass (kg)'].quantile(0.75)
# Calculate the Interquartile Range (IQR)
IQR = Q3 - Q1
# Values below this are considered outliers
lower_bound = Q1 - 1.5 * IQR
# Values below this are considered outliers
upper_bound = Q3 + 1.5 * IQR
# This removes the outliers from the dataset
df_clean = df[(df['Payload Mass (kg)'] >= lower_bound) & (df['Payload Mass (kg)'] <= upper_bound)]

print('Toatl rows before removing outliters', len(df))
print("Rows after removing outliers:", len(df_clean))
```

### Coorelation graph:
A correlation graph, most commonly a scatterplot, is a visual representation of the relationship between two variables

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Convert the 'Launch Date' column into datetime format
df['Launch Date'] = pd.to_datetime(df['Launch Date'])
# Convert Launch Date into numeric format 
df['Launch Date Numeric'] = df['Launch Date'].astype(int) / 10**9  
# Select only numeric columns for correlation
numeric_cols = ['Payload Mass (kg)', 'Launch Date Numeric']
# calcualtion
corr_matrix = df[numeric_cols].corr()

# our rquirements for graph
plt.figure(figsize=(8,5))
sns.heatmap(
    corr_matrix,
    annot=True,
    cmap='plasma',
    linewidths=0.5
)

plt.show()
```

### Hypothesis testing:
Hypothesis testing is a statistical method used to check whether a claim or assumption about a population is true or not.

```python
import pandas as pd
import numpy as np
from scipy import stats
# Create a small dataset
data = {
    'Flight Number': ['F1-1','F1-2','F1-3','F1-3','F1-3'],
    'Payload Mass (kg)': [19.5, 45.0, 123.0, 8.0, 34.9]
}
df = pd.DataFrame(data)
sample = df['Payload Mass (kg)'].dropna()
# Find the sample size
n = len(sample)
# calculate sample mean
sample_mean = sample.mean()
# Calculate the sample standard deviation
sample_std = sample.std(ddof=1)  
# print mean,sd,and length of sample
print("Sample Mean:", sample_mean)
print("Sample Std Dev:", sample_std)
print("Sample Size:", n)

# Assume a population mean value
population_mean = 15
t_stat, p_value = stats.ttest_1samp(sample, population_mean)
# result of t statistics and z test
print("\nT-statistic:", t_stat)
print("P-value:", p_value)
# Set significance level
alpha = 0.05
# Make decision based on p-value
if p_value < alpha:
    print("Reject H0: Payload Mass mean is significantly different from 15 kg.")
else:
    print("Fail to reject H0: Payload Mass mean is NOT significantly different from 15 kg.")
```
